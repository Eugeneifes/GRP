```python
import pandas as pd
import matplotlib.pyplot as plt
from math import sin, cos, sqrt, atan2, radians

data = pd.read_csv('01_test_data.csv')

#approximate flat distance between two points in km
def flat_dist(lat1, lon1, lat2, lon2):
    
    #approximate radius of earth in km
    R = 6371.0
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    
    x = (lon2 - lon1)*cos(0.5*(lat2+lat1))
    y = lat2 - lat1
    distance = R*sqrt(x*x + y*y)
    return distance


#approximate convex distance between two points in km
def convex_dist(lat1, lon1, lat2, lon2):
    
    #approximate radius of earth in km
    R = 6371.0
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    
    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat/2)**2 + cos(lat1)*cos(lat2)*sin(dlon/2)**2
    c = 2*atan2(sqrt(a), sqrt(1 - a))

    distance = R*c
    return distance
    
for index, row in data.head(100).iterrows():
    print(flat_dist(row['begin_lat'], row['begin_lon'], row['end_lat'], row['end_lon']))
    print(convex_dist(row['begin_lat'], row['begin_lon'], row['end_lat'], row['end_lon']))
    print('\n')
```
